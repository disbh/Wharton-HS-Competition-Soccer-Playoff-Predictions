```python
import os
import pandas as pd

# Load the data
data_path = os.path.join(os.path.dirname(os.getcwd()), 'data', 'NSL_regular_season_data_2.csv')
print(data_path)
data = pd.read_csv(data_path)
```

```python
# Find the rankings of the teams
d = {}
for i in range(len(data)):
    winner, loser, tie1, tie2 = '', '', '', ''
    if data['HomeScore'][i] > data['AwayScore'][i]:
        winner = data['HomeTeam'][i]
        loser = data['AwayTeam'][i]
    elif data['HomeScore'][i] < data['AwayScore'][i]:
        winner = data['AwayTeam'][i]
        loser = data['HomeTeam'][i]
    else:
        tie1 = data['HomeTeam'][i]
        tie2 = data['AwayTeam'][i]
    if winner:
        if winner not in d:
            d[winner] = [1, 0, 0]
        else:
            d[winner][0] += 1
    if loser:
        if loser not in d:
            d[loser] = [0, 1, 0]
        else:
            d[loser][1] += 1
    if tie1:
        if tie1 not in d:
            d[tie1] = [0, 0, 1]
        else:
            d[tie1][2] += 1
    if tie2:
        if tie2 not in d:
            d[tie2] = [0, 0, 1]
        else:
            d[tie2][2] += 1

# Create a dataframe from the dictionary
df = pd.DataFrame.from_dict(d, orient='index')
df.columns = ['Wins', 'Losses', 'Ties']
df['WinPct'] = df['Wins'] / (df['Wins'] + df['Losses'] + df['Ties'])

# Order the data alphabetically
df = df.sort_index()
df
```

```python
# Save the data
df.to_csv(os.path.join(os.path.dirname(os.getcwd()), 'data', 'NSL_rankings.csv'))
```
